```python
# classify labels into two classes


import pandas as pd
import os
import matplotlib.pyplot as plt
import shutil

ROOT = '/content/drive/MyDrive/mitosis_detector_BTTAI/BTTAI_Images/patches/patches'
class_1 = '/content/drive/MyDrive/mitosis_detector_BTTAI/BTTAI_Images/patches/1'
class_2 = '/content/drive/MyDrive/mitosis_detector_BTTAI/BTTAI_Images/patches/2'

#os.makedirs(class_1, exist_ok=True)
#os.makedirs(class_2, exist_ok=True)

annot_df = pd.read_csv('annotation.csv')

#print("Annotation data:")
#display(annot_df.head())

#os.mkdir("mitotic_figure")
#os.mkdir("not mitotic")

imgs_with_bbox = annot_df[['category_id', 'image_id', 'bbox_id']].copy()
#display(imgs_with_bbox)

for img_ID, group in imgs_with_bbox.groupby("image_id"):
  img_dir = os.path.join(ROOT, str(img_ID))

  for _, row in group.iterrows():
    filename = f"{img_ID}_{row.bbox_id}.jpeg"
    src_path = os.path.join(ROOT, str(img_ID), filename)

    category = row['category_id']

    if category == 1:
      print(f"copying {filename} with category {category} to {class_1}")
      shutil.copy(src_path, class_1)

    else:
      print(f"copying {filename} with category {category} to {class_2}")
      shutil.copy(src_path, class_2)


```
