```python
FN_imgs = os.listdir(FN_PATH)
FP_imgs = os.listdir(FP_PATH)

FN_types = []
FP_types = []

for s1 in FN_imgs:
  id1 = s1.partition("_")[0]
  if len(id1) == 2:
    id1 = "0" + id1
  elif len(id1) == 1:
    id1 = "00" + id1
  cond1 = df['file_name'] == id1 + ".tiff"
  idx1 = df.index[cond1]
  name1 = df.loc[idx1, 'tumor_type'].iloc[0]
  FN_types.append((s1, name1))

for s in FP_imgs:
  id = s.partition("_")[0]
  if len(id) == 2:
    id = "0" + id
  elif len(id) == 1:
    id = "00" + id

  cond = df['file_name'] == id + ".tiff"
  idx = df.index[cond]
  name = df.loc[idx, 'tumor_type'].iloc[0]
  FP_types.append((s, name))
```

```python
IMG_CSV = "/content/drive/MyDrive/mitosis_detector_BTTAI/images.csv"
df = pd.read_csv(IMG_CSV)
FN_imgs = os.listdir(FN_PATH)
FP_imgs = os.listdir(FP_PATH)
FN_types = []
FP_types = []

for str_fn in FN_imgs:
  id_fn = str_fn.partition("_")[0]
  if len(id_fn) == 2:
    id_fn_f = "0" + id_fn
  elif len(id_fn) == 1:
    id_fn_f = "00" + id_fn
  else:
    id_fn_f = id_fn
  cond_fn = df['file_name'] == id_fn_f + ".tiff"
  idx_fn = df.index[cond_fn]
  name_fn = df.loc[idx_fn, 'tumor_type'].iloc[0]
  FN_types.append((str_fn, name_fn))

for str_fp in FP_imgs:
  id_fp = str_fp.partition("_")[0]
  if len(id_fp) == 2:
    id_fp_f = "0" + id_fp
  elif len(id_fp) == 1:
    id_fp_f = "00" + id_fp
  else:
    id_fp_f = id_fp

  cond_fp = df['file_name'] == id_fp_f + ".tiff"
  idx_fp = df.index[cond_fp]
  name_fp = df.loc[idx_fp, 'tumor_type'].iloc[0]
  FP_types.append((str_fp, name_fp))


  print("distribution of tumor types in images slides:")

  print(df['tumor_type'].value_counts())

print("Cell Types for False Negatives:")
print("-------------------------------")
for item in FN_types:
  print(item)

print("Cell Types for False Positives:")
print("-------------------------------")
for item in FP_types:
  print(item)

def print_groups(samples):
    groups = {}
    for img_id, category in samples:
        groups.setdefault(category, []).append(img_id)

    for category, ids in groups.items():
        print(f"{category} ({len(ids)} images): {ids}")

print("False Negatives:")
print_groups(FN_types)
print("\nFalse Positives:")
print_groups(FP_types)
```
